```python
#optimized retrieval 
import urllib.request
import csv
import pandas as pd
import json
import re

# categories with associated offence code
larceny = [612, 617, 619, 621, 611, 631, 612, 622, 632]
manslaughter = [123, 121, 111, 121, 122, 123, 124, 125]
driving_under_influence = [2102, 2101]
rape_sexual_assault = [254, 271, 244, 251, 241, 243, 261, 252, 253, 1704, 222, 230, 236, 211, 232, 233, 224, 213, 223, 231, 237, 212, 234, 235, 1720, 1703]
robbery = [301, 311, 351, 361, 371, 381, 301, 302, 304, 303, 305, 306, 307, 308, 309, 310, 311, 312, 313, 314, 315, 316, 317, 318, 319, 320, 321, 322, 323, 324, 333, 334, 335, 336, 337, 338, 339, 340, 341, 342, 434, 344, 345, 346, 347, 348, 349, 350, 351, 352, 353, 354, 355, 356, 357, 358, 359, 360, 361, 362, 363, 364, 373, 374, 375, 376, 377, 378, 379, 380]
assault_battery = [801, 802, 423, 413, 1702, 401, 402, 403, 404, 411, 413, 421, 422, 423, 424, 431, 432, 433, 1727, 1726, 1725, 1702]
kidnapping = [2511, 2622, 2611]
burglary = [541, 540, 562, 561, 542, 521, 522, 560]
bombs_explosives = [2648, 2618, 3123, 3107]
harassment = [2670, 2629]
posession_drugs = [1810, 1847, 1840, 1843, 1841, 1848, 1850]
weapons = [3016, 3017, 1501, 1503, 1502, 1504, 1510, 3119]
arson = [900, 901, 902, 905, 906, 907, 920, 930]
breaking_entering = [520, 521, 522, 527, 510, 511, 512, 517, 540, 541, 542, 543, 547, 530, 531, 532, 537]
child_abuse_endangerment = [2003, 2005]
disorderly_conduct = [2405, 2403]
public_drinking = [2646]


def offense_type(code):
    if code in larceny:
        return 'Larceny'
    elif code in manslaughter:
        return 'Manslaughter'
    elif code in driving_under_influence:
        return 'Driving Under Influence'
    elif code in rape_sexual_assault:
        return 'Rape/Sexual Assault'
    elif code in robbery:
        return 'Robbery'
    elif code in assault_battery:
        return 'Assault/Battery'
    elif code in kidnapping:
        return 'Kidnapping'
    elif code in burglary:
        return 'Burglary'
    elif code in bombs_explosives:
        return 'Bombs/Explosives'
    elif code in harassment:
        return 'Harassment'
    elif code in posession_drugs:
        return 'Possession'
    elif code in weapons:
        return 'Weapons'
    elif code in arson:
        return 'Arson'
    elif code in breaking_entering:
        return 'Breaking and Entering'
    elif code in child_abuse_endangerment:
        return 'Child Abuse/Endangerment'
    elif code in disorderly_conduct:
        return 'Disorderly Conduct'
    elif code in public_drinking:
        return 'Public Drinking'
    else:
        return None
    
# process all in the dataframe 
def process(R, f):
    return [f(r) for r in R]

# filter results based on relevant types to the study
def filterResults(row):
    try:
        if offense_type(int(row['offense_code'])) and row['offense_code'] and row['district'] and row['hour'] and row['location']:
            return {'offense_type': offense_type(int(row['offense_code'])), 'district': row['district'], 'hour': row['hour'], 'location': row['location']['coordinates']}
        else:
            return None
    except KeyError:
        return None


if __name__ == "__main__":
    # retrieve the data
    url = 'https://data.cityofboston.gov/resource/29yf-ye7n.json'
    response = urllib.request.urlopen(url).read().decode("utf-8")
    data = json.loads(response)

    # filter through results first
    crime_data = process(data, filterResults)
    crime_data = [r for r in crime_data if r] #removes None values
    
    # save to a pandas dataframe
    crime_data = pd.DataFrame.from_dict(crime_data)
    print(crime_data)
    
```
