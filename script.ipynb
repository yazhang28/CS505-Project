```python
import urllib.request
import csv
import pandas as pd

# fetching the csv 
response = urllib.request.urlopen('https://data.cityofboston.gov/api/views/fqn4-4qap/rows.csv?accessType=DOWNLOAD&bom=true')
data = response.read()
# Save the string to a file
csvstr = str(data).strip("b'")
# saving the csv
lines = csvstr.split("\\n")
d = open("crime_incident_report_august_2015_to_date.csv", "w")
for line in lines:
    d.write(line + "\n")

#reading in csv to create dataframe
df = pd.read_csv('crime_incident_report_august_2015_to_date.csv')

# removing irrelevant columns and null values
cols = [0,2,3,5,6,7,9,10,11,12,13,14,15]
df.drop(df.columns[cols],axis=1,inplace=True)

df = df[df.DISTRICT.notnull()]
df = df[df.Location != '(0E-8, 0E-8)']
df
```
